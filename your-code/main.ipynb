```python
# Your code here

# Function to plot the heatmap
def plot_heatmap(corr_matrix, title):
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
    plt.title(title)
    plt.show()

# Function to remove highly correlated features
def remove_highly_correlated_features(numeric_df, threshold=0.90):
    # Calculate the correlation matrix
    corr_matrix = numeric_df.corr().abs()
    
    # Plot initial heatmap
    plot_heatmap(corr_matrix, 'Initial Correlation Matrix')

    # Select the upper triangle of the correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    # Iterate over columns to find pairs with correlation greater than the threshold
    while True:
        # Find index of feature columns with correlation greater than the threshold
        to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
        
        if not to_drop:
            break
        
        # Drop the first column found with high correlation
        col_to_remove = to_drop[0]
        print(f"Removing column: {col_to_remove} due to high correlation.")
        numeric_df = numeric_df.drop(columns=[col_to_remove])
        
        # Recalculate the correlation matrix
        corr_matrix = numeric_df.corr().abs()
        upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    # Plot final heatmap
    plot_heatmap(corr_matrix, 'Final Correlation Matrix (No High Correlations)')
    
    return numeric_df

# Assuming 'numeric_df' is your DataFrame with numerical features
websites = remove_highly_correlated_features(websites)
```

#### Before you think of your own solution, don't read the instructions that come next.

```python
# Your code here

# print the number of unique values in SERVER
print("Number of unique values in SERVER: ", websites['SERVER'].nunique())

# consolidate SERVER values into broader categories
def consolidate_server(server):
    if pd.isna(server):  # Handle missing values
        return 'Other'
    elif 'microsoft' in server.lower():  # Microsoft servers (including Microsoft-IIS)
        return 'Microsoft'
    elif 'apache' in server.lower():  # Apache servers
        return 'Apache'
    elif 'nginx' in server.lower():  # nginx servers
        return 'nginx'
    else:  # All other servers
        return 'Other'

# apply the consolidation function to the SERVER column
websites['SERVER_CONSOLIDATED'] = websites['SERVER'].apply(consolidate_server)

# print the unique values in the consolidated SERVER column
print("Unique values in SERVER_CONSOLIDATED: ", websites['SERVER_CONSOLIDATED'].unique())
```

Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`.
